```python
from eda import *
from gradientBoosting import *
import xgboost as xgb
```

# Preprocessing

```python
file_path = 'Data/exchange_v2.csv'
train_df, test_df = load_train_test_data(
    file_name=file_path, 
    currency='BTC-USD',
    train_size=0.85
)
```

```python
train_df, charts = preprocess(train_df)
```

# Gradient Boosting

```python
#define the input features for the gradient boosting, has to be a list ['feature1', 'feature2', 'feature3']
features = ['date']
#define the target for the training i.e 'Close', 'mov7', 'mov30'
target_variable = 'mov7'
```

```python
model, trainPredictions, testPredictions, trainR2, testR2 = gradient_boosting(train_df,test_df,features,target_variable)

print(f"The R2 score for the train data is: {trainR2}")
print(f"The R2 score for the test data is: {testR2}")
```

```python
create_plot(train_df[target_variable], trainPredictions)
```

```python
create_plot(test_df['Close'], testPredictions)
```

```python
#Plot the feature importance
xgb.plot_importance(model)
```
